### Document Loader

```python
#pip install -qU langchain-community pypdf
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
file_path = "D:\\Data Science_ML\\LangGraph\\rag-with-langgraph\\llm-book.pdf"
loader = PyPDFLoader(file_path)
```

```python
docs = loader.load()
docs[0].page_content
```

```python
import pprint

pprint.pp(docs[1].page_content)
```

### Text Splitters

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=0)
texts = text_splitter.split_documents(docs)
```

```python
texts
```

### Embeddings


```python
#pip install langchain_openai
```

```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
#pip install faiss-cpu
```

```python
embeddings = OpenAIEmbeddings()

vector_store = FAISS.from_documents(
        texts, 
        embedding=embeddings
        )

vector_store.save_local("rag_embeddings")
```

```python
vector_store = FAISS.load_local(
    "rag_embeddings",
    embeddings,
    allow_dangerous_deserialization=True
)
```

```python
vector_store
```

### Retrievers

```python
vector_store.similarity_search("Stages of building and using LLMs")
```

### Augmentation

```python
retrieved_docs = vector_store.similarity_search("Stages of building and using LLMs")

context = "\n\n".join(doc.page_content for doc in retrieved_docs)

context
```
